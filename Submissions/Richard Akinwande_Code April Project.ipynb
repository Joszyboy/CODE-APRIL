```python
#DATA EXPLORATION
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

Titanic = 'tested.csv'

df = pd.read_csv(Titanic)

df.head
```

```python
df.head()
```

```python
df.describe()
```

```python
# Examine passenger profiles
# Let's visualize the distribution of passenger age
plt.figure(figsize=(8, 6))
df['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()


# Let's visualize the distribution of passenger base on gender
plt.figure(figsize=(6, 4))
df['Sex'].hist(bins=20, color='red', edgecolor='black')
plt.title('Distribution of Passenger Base on Sex')
plt.xlabel('Frequency')
plt.ylabel('Sex')
plt.grid(False)
plt.show()
```

```python
# Examine survival rates
# Let's calculate the overall survival rate
survival_rate = df['Survived'].mean() * 100
print("\nOverall Survival Rate: {:.2f}%".format(survival_rate))


# Survival rate by gender
survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100
print("\nSurvival Rate by Gender:")
print(survival_by_gender)


plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

```

```python
# Survival rate by passenger class
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival Rate by Passenger Class:")
print(survival_by_class)


survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(survival_by_class, labels=survival_by_class.index, autopct='%1.1f%%', startangle=140)
plt.title('Survival Rate by Passenger Class')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


# Factors affecting outcomes
# Let's explore the relationship between passenger class and survival
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival rate by age group
# Let's categorize passengers into different age groups
bins = [0, 18, 30, 50, 80]
labels = ['0-17', '18-29', '30-49', '50+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
survival_by_age_group = df.groupby('AgeGroup')['Survived'].mean() * 100
print("\nSurvival Rate by Age Group:")
print(survival_by_age_group)


# Let's explore the relationship between age group and survival
plt.figure(figsize=(8, 6))
sns.barplot(x='AgeGroup', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket'], axis=1)

# Convert categorical features to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['Sex', 'Embarked'])

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Plot scatter plot of Age vs. Fare colored by survival
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title('Scatter Plot of Age vs. Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()

# Plot box plot of Age across survival categories
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Box Plot of Age Across Survival Categories')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Check the columns present in the DataFrame
print(df.columns)

# Plot count plots of Passenger Class and Sex across survival categories
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Count Plot of Passenger Class Across Survival Categories')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')

if 'Sex' in df.columns:
    plt.subplot(1, 2, 2)
    sns.countplot(data=df, x='Sex', hue='Survived')
    plt.title('Count Plot of Sex Across Survival Categories')
    plt.xlabel('Sex')
    plt.ylabel('Count')
    plt.legend(title='Survived', loc='upper right')

plt.tight_layout()
plt.show()
```
